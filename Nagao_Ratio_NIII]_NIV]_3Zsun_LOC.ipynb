```python
fig = plt.figure(figsize=(8,6)) 

# from matplotlib import rc
# rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# ## for Palatino and other serif fonts use:
# #rc('font',**{'family':'serif','serif':['Palatino']})
# rc('text', usetex=True)

import matplotlib
matplotlib.rcParams['text.usetex'] = False
#matplotlib.rcParams['text.latex.unicode'] = True

# plt.plot(N, (n3/o3)[0:11],lw=2, label=r'$\log n_H = 10$, $\log \phi = 19.5$')

for i in range(25):
    plt.plot(N, np.split((n3/o3), 25)[i],lw=5, alpha=0.1, c='C0')
    #plt.plot(N, np.mean(np.split((n3/o3), 25)[i]),lw=2, alpha=0.5, c='grey')
    plt.ylabel('NIII] / OIII]', fontsize=20)
    plt.xlabel('N scale factor', fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.ylim(0.15,1.8)
    plt.tick_params(direction='in', which='both')
    #plt.yscale('log')
    plt.grid('True',which="both")
    plt.title(r'$Z_{\alpha}=3\,Z_{\alpha,\odot}$', fontsize=20)
    
plt.plot(N, n3o3_m,lw=3, alpha=1, c='C1', label='LOC average')
plt.legend(fontsize=15)
plt.savefig('n3o3_3xsloar_alpha.png', dpi=300, facecolor='white', edgecolor='white')
plt.show();
```
