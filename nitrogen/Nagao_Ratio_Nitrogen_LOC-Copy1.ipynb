```python
fig = plt.figure(figsize=(6,6)) 
plt.rc('axes', linewidth=2)
font = {'family' : 'serif',
        'size'   : 20,}
plt.rc('font', **font)
plt.plot(N, n3/o3, label=r'n10, $\phi$19',lw=2)
plt.plot(N, n31/o31, label=r'n10, $\phi$20',lw=2)
plt.plot(N, n32/o32, label=r'n11, $\phi$19',lw=2)
plt.plot(N, n33/o33, label=r'n11, $\phi$20',lw=2)

plt.plot(N, n3o3_loc, label=r'LOC average',lw=2, ls='--', c='k')

plt.ylabel('N III] / O III]', fontsize=15)
plt.xlabel(r'(N/O)/(N/O)$_{\odot}$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)
#plt.xscale('log')
plt.grid('True')
plt.title(r'$Z=3Z_{\odot}$', fontsize=20)
plt.tick_params(axis='both', which='major',length=5,width=2,direction='in')
plt.tick_params(axis='both', which='minor',length=2,width=1,direction='in')
plt.savefig('m3_nitrogen_4LOC.png', dpi=300, facecolor='white', edgecolor='white');
```
